```python
import pandas as pd
import datetime
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
rating_column_names = 'user id | item id | rating | timestamp'.split(' | ')
user_column_names = 'user id | age | gender | occupation | zip code'.split(' | ')
item_column_names = 'movie id | movie title | release date | video release date | IMDb URL | unknown | Action | Adventure | Animation | Children | Comedy | Crime | Documentary | Drama | Fantasy | Film-Noir | Horror | Musical | Mystery | Romance | Sci-Fi | Thriller | War | Western'.split(' | ')

rating_data =  pd.read_csv(r'D:\Exercises\ML Lab\ml-100k\u.data', sep='\t',header = None, names = rating_column_names)
user_data =  pd.read_csv(r'D:\Exercises\ML Lab\ml-100k\u.user', sep='|',header = None, names = user_column_names)
item_data =  pd.read_csv(r'D:\Exercises\ML Lab\ml-100k\u.item', sep='|',header = None, names = item_column_names,encoding = "ISO-8859-1")
rating_data['timestamp'] = pd.to_datetime(rating_data['timestamp'],unit='s')
item_data['release year'] = pd.DatetimeIndex(item_data['release date']).year
```

### Distribution of Ratings based on occupations

```python
#rating_data.loc[user_data['gender'] == 'M']
occ= user_data['occupation'].unique()
rat_dis={}
for oc in occ:
    ids = list(user_data.loc[(user_data["occupation"] == oc), "user id"])
    rat_dis[oc] = rating_data.loc[rating_data['user id'].isin(ids),"rating"]
```

```python
fig = plt.figure(1, figsize=(16, 6))
# Create an axes instance
ax = fig.add_subplot(111)
#ax.set_xticklabels(occ)
ax.get_xaxis().tick_bottom()
ax.get_yaxis().tick_left()
# Create the boxplot
bp = ax.boxplot(rat_dis.values())
#ax.set_xticklabels(occ)
plt.xticks(np.arange(1,len(occ),1),occ, rotation=55)
plt.ylabel('Rating', fontsize = 20)
plt.xlabel('Occupation', fontsize = 20)
plt.title("Average Rating based on Profession", fontsize = 20)
plt.grid()
```

### Distribution of rating for each genre

```python
genres='Action | Adventure | Animation | Children | Comedy | Crime | Documentary | Drama | Fantasy | Film-Noir | Horror | Musical | Mystery | Romance | Sci-Fi | Thriller | War | Western'.split(' | ')
avg_rat_g={}
for genre in genres:
    avg_rat_g[genre]=np.mean(rating_data['rating'].loc[rating_data['item id'].isin(list(item_data['movie id'].loc[item_data[str(genre)]==1]))])-3

fig = plt.figure(1, figsize=(20,13))
ax2 = fig.add_subplot(2,1,2)
plt.xticks(rotation=75, fontsize = 15)
plt.yticks(fontsize = 15)
ax2.bar(avg_rat_g.keys(),avg_rat_g.values(),align='center', color='b') # A bar chart
plt.xlabel('Genre', fontsize = 20)
plt.grid()
plt.title("Avg Rating for each genre", fontsize = 20)
plt.ylabel('Avg Rating(3+)', fontsize = 20)
```

### Distribution of rating for each genre in males

```python
genres='Action | Adventure | Animation | Children | Comedy | Crime | Documentary | Drama | Fantasy | Film-Noir | Horror | Musical | Mystery | Romance | Sci-Fi | Thriller | War | Western'.split(' | ')
avg_rat_g_m={}
for genre in genres:
    iid_array=item_data['movie id'].loc[item_data[str(genre)]==1]
    uid_array=user_data['user id'].loc[user_data['gender'].isin([str('M')])]
    avg_rat_g_m[genre]=(np.mean(rating_data['rating'].loc[(rating_data['user id'].isin(uid_array)) & rating_data['item id'].isin(iid_array)]))-3
    

fig = plt.figure(1, figsize=(20,13))
ax2 = fig.add_subplot(2,1,2)
plt.xticks(rotation=75, fontsize = 15)
plt.yticks(fontsize = 15)
ax2.bar(avg_rat_g_m.keys(),avg_rat_g_m.values(),align='center', color='b') # A bar chart
plt.xlabel('Genre', fontsize = 20)
plt.title("Avg Rating for each genre Male", fontsize = 20)
plt.ylabel('Avg Rating (3+)', fontsize = 20)
plt.grid()
```

### Distribution of rating for each genre in Females

```python
genres='Action | Adventure | Animation | Children | Comedy | Crime | Documentary | Drama | Fantasy | Film-Noir | Horror | Musical | Mystery | Romance | Sci-Fi | Thriller | War | Western'.split(' | ')
avg_rat_g_f={}
for genre in genres:
    iid_array=item_data['movie id'].loc[item_data[str(genre)]==1]
    uid_array=user_data['user id'].loc[user_data['gender'].isin([str('F')])]
    avg_rat_g_f[genre]=(np.mean(rating_data['rating'].loc[(rating_data['user id'].isin(uid_array)) & rating_data['item id'].isin(iid_array)]))-3
    

fig = plt.figure(1, figsize=(20,13))
ax2 = fig.add_subplot(2,1,2)
plt.xticks(rotation=75, fontsize = 15)
plt.yticks(fontsize = 15)
ax2.bar(avg_rat_g_f.keys(),avg_rat_g_f.values(),align='center', color='b') # A bar chart
plt.xlabel('Genre', fontsize = 20)
plt.title("Avg Rating for each genre Female", fontsize = 20)
plt.ylabel('Avg Rating (3+)', fontsize = 20)
plt.grid()
```
